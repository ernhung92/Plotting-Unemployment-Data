```python
import pandas as pd

unrate = pd.read_csv("UNRATE.csv")
# Converting date data into datetime
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
unrate.head(12)
```

```python
import matplotlib.pyplot as plt

# Creating an empty plot
plt.plot()
plt.show()
```

```python
# Creating a plot for unemployment rate in 1948
first_twelve = unrate[0:12]
plt.plot(first_twelve["DATE"], first_twelve["UNRATE"])
plt.show()
```

```python
# From the previous plot, it looks like the x-axis tick labels are so close to each other.
plt.plot(first_twelve["DATE"], first_twelve["UNRATE"])
# We rotate the x-axis 90 degrees
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(first_twelve['DATE'], first_twelve['UNRATE'])
plt.xticks(rotation=90)
# Adding xlabels and a title
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
```

```python
ax1.plot(first_twelve["DATE"], first_twelve["UNRATE"], c='red')
second_year = unrate[12:24]
ax2.plot(second_year["DATE"], second_year["UNRATE"], c='blue')
plt.show()
```

```python
# Creating visualization for many years
fig2 = plt.figure(figsize=(12,12))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    axe = fig2.add_subplot(5,1,i+1)
    start_index = i * 12
    end_index = (i+1) * 12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    axe.plot(subset["DATE"],subset["UNRATE"], c=colors[i], label=label)
    plt.legend(loc='upper left')
    plt.xlabel('Month, Integer')
    plt.ylabel('Unemployment Rate, Percent')
    plt.title('Monthly Unemployment Trends, 1948-1952')
    
plt.show()
```
