```python
import pandas as pd

reviews = pd.read_csv("fandango_score_comparison.csv")
column_names = ["FILM", "RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
norm_reviews = reviews[column_names]
norm_reviews.head(1)
```

```python
import matplotlib.pyplot as plt
from numpy import arange
bar_heights = norm_reviews[column_names].iloc[0].values
print(bar_heights)
bar_positions = arange(5) + 0.75
print(bar_positions)
```

```python
fig, ax = plt.subplots()
ax.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])
ax.set_xlabel('Fandango')
ax.set_ylabel('Rotten Tomatoes')
plt.show()
```

```python
fig2 = plt.figure(figsize=(12,12))
ax1 = fig2.add_subplot(2,1,1)
ax2 = fig2.add_subplot(2,1,2)

ax1.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")

ax2.scatter(norm_reviews["RT_user_norm"], norm_reviews["Fandango_Ratingvalue"])
ax2.set_xlabel("Rotten Tomatoes")
ax2.set_ylabel("Fandango")
plt.show()
```

```python
fig3 = plt.figure(figsize=(5,10))
ax11 = fig3.add_subplot(3,1,1)
ax22 = fig3.add_subplot(3,1,2)
ax33 = fig3.add_subplot(3,1,3)

ax11.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax11.set_xlabel("Fandango")
ax11.set_ylabel("Rotten Tomatoes")
ax11.set_xlim(0,5)
ax11.set_ylim(0,5)

ax22.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["Metacritic_user_nom"])
ax22.set_xlabel("Fandango")
ax22.set_ylabel("Metacritic")
ax22.set_xlim(0,5)
ax22.set_ylim(0,5)

ax33.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["IMDB_norm"])
ax33.set_xlabel("Fandango")
ax33.set_ylabel("IMDB")
ax33.set_xlim(0,5)
ax33.set_ylim(0,5)

plt.show()
```

```python
# By looking at the scatter plots, we can conclude that IMDB and Fandango are the most similar
# Rotten Tomatoes and Metacritic have a weak but positive correlation 
```

```python
fandango_distribution = norm_reviews["Fandango_Ratingvalue"].value_counts().sort_index()
print(fandango_distribution)

imdb_distribution = norm_reviews["IMDB_norm"].value_counts().sort_index()
print(imdb_distribution)
```

```python
fig4, ax40 = plt.subplots()
ax40.hist(norm_reviews["Fandango_Ratingvalue"], range=(0,5))
plt.show()
```

```python
fig5 = plt.figure(figsize=(5,20))
ax51 = fig5.add_subplot(4,1,1)
ax52 = fig5.add_subplot(4,1,2)
ax53 = fig5.add_subplot(4,1,3)
ax54 = fig5.add_subplot(4,1,4)

ax51.hist(norm_reviews['Fandango_Ratingvalue'], bins=20, range=(0, 5))
ax51.set_title('Distribution of Fandango Ratings')
ax51.set_ylim(0, 50)

ax52.hist(norm_reviews['RT_user_norm'], 20, range=(0, 5))
ax52.set_title('Distribution of Rotten Tomatoes Ratings')
ax52.set_ylim(0, 50)

ax53.hist(norm_reviews['Metacritic_user_nom'], 20, range=(0, 5))
ax53.set_title('Distribution of Metacritic Ratings')
ax53.set_ylim(0, 50)

ax54.hist(norm_reviews['IMDB_norm'], 20, range=(0, 5))
ax54.set_title('Distribution of IMDB Ratings')
ax54.set_ylim(0, 50)

plt.show()
```

```python
fig6, ax60 = plt.subplots()
ax60.boxplot(norm_reviews['RT_user_norm'])
ax60.set_xticklabels(['Rotten Tomatoes'])
ax60.set_ylim(0, 5)
plt.show()
```

```python
# From the box plot we generated using Rotten Tomatoes ratings, we can conclude that the bottom 25% of user ratings range from around 1 to 2.5, the top 25% of of user ratings range from around 4 to 4.6 
```

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']

fig7 ,ax70 = plt.subplots()
ax70.boxplot(norm_reviews[num_cols].values)
ax70.set_xticklabels(num_cols, rotation=90)
ax70.set_ylim(0,5)
plt.show()
```
